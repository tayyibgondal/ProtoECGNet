```python
import pandas as pd

# Load the CSV files
resnet18_df = pd.read_csv('resnet18.csv')
ppnet_df = pd.read_csv('ppnet.csv')

# Calculate the average metrics across all classes for each model
resnet18_metrics = {
    'accuracy': resnet18_df['accuracy'].mean(),
    'auroc': resnet18_df['auroc'].mean(),
    'f1_score': resnet18_df['f1_score'].mean(),
}

ppnet_metrics = {
    'accuracy': ppnet_df['accuracy'].mean(),
    'auroc': ppnet_df['auroc'].mean(),
    'f1_score': ppnet_df['f1_score'].mean(),
}

# Display the results
print("ResNet18 Average Metrics")
for metric, value in resnet18_metrics.items():
    print(f"{metric}: {value:.4f}")

print("\nPPNET Average Metrics")
for metric, value in ppnet_metrics.items():
    print(f"{metric}: {value:.4f}")

# Compare the models
comparison = {metric: resnet18_metrics[metric] > ppnet_metrics[metric] for metric in resnet18_metrics}

print("\nModel Comparison (True means ResNet18 is better):")
for metric, is_resnet18_better in comparison.items():
    print(f"{metric}: {is_resnet18_better}")
```

```python
import pandas as pd

# Load the CSV files
resnet18_df = pd.read_csv('resnet18.csv')
ppnet_df = pd.read_csv('ppnet.csv')

# Calculate the average metrics across all classes for each model
resnet18_metrics = {
    'accuracy': resnet18_df['accuracy'].mean(),
    'auroc': resnet18_df['auroc'].mean(),
    'f1_score': resnet18_df['f1_score'].mean(),
}

ppnet_metrics = {
    'accuracy': ppnet_df['accuracy'].mean(),
    'auroc': ppnet_df['auroc'].mean(),
    'f1_score': ppnet_df['f1_score'].mean(),
}

# Display the results
print("ResNet18 Average Metrics")
for metric, value in resnet18_metrics.items():
    print(f"{metric}: {value:.4f}")

print("\nPPNET Average Metrics")
for metric, value in ppnet_metrics.items():
    print(f"{metric}: {value:.4f}")

# Compare the models
comparison = {metric: resnet18_metrics[metric] > ppnet_metrics[metric] for metric in resnet18_metrics}

print("\nModel Comparison (True means ResNet18 is better):")
for metric, is_resnet18_better in comparison.items():
    print(f"{metric}: {is_resnet18_better}")

# Calculate the percentage improvement of PPNET over ResNet18
percentage_improvement = {
    metric: ((ppnet_metrics[metric] - resnet18_metrics[metric]) / resnet18_metrics[metric]) * 100
    for metric in resnet18_metrics
}

print("\nPercentage Improvement of PPNET over ResNet18:")
for metric, improvement in percentage_improvement.items():
    if improvement > 0:
        print(f"{metric}: {improvement:.2f}% better")
    else:
        print(f"{metric}: {-improvement:.2f}% worse")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV files
resnet18_df = pd.read_csv('resnet18.csv')
ppnet_df = pd.read_csv('ppnet.csv')

# Calculate the average metrics across all classes for each model
resnet18_metrics = {
    'accuracy': resnet18_df['accuracy'].mean(),
    'auroc': resnet18_df['auroc'].mean(),
    'f1_score': resnet18_df['f1_score'].mean(),
}

ppnet_metrics = {
    'accuracy': ppnet_df['accuracy'].mean(),
    'auroc': ppnet_df['auroc'].mean(),
    'f1_score': ppnet_df['f1_score'].mean(),
}

# Display the results
print("ResNet18 Average Metrics")
for metric, value in resnet18_metrics.items():
    print(f"{metric}: {value:.4f}")

print("\nPPNET Average Metrics")
for metric, value in ppnet_metrics.items():
    print(f"{metric}: {value:.4f}")

# Plotting the metrics for both models
metrics = ['accuracy', 'auroc', 'f1_score']
resnet18_values = [resnet18_metrics[metric] for metric in metrics]
ppnet_values = [ppnet_metrics[metric] for metric in metrics]

x = range(len(metrics))
bar_width = 0.35

plt.figure(figsize=(10, 6))
plt.bar(x, resnet18_values, width=bar_width, label='ResNet18', color='steelblue')
plt.bar([p + bar_width for p in x], ppnet_values, width=bar_width, label='PPNET', color='darkorange')

plt.xlabel('Metrics', fontsize=14)
plt.ylabel('Values', fontsize=14)
plt.title('Comparison of Average Metrics between ResNet18 and PPNET', fontsize=16)
plt.xticks([p + bar_width / 2 for p in x], metrics)
plt.legend(fontsize=12)

# Remove top and right spines for a cleaner look
sns.despine()

# Save the combined plot as an image file
plt.savefig('model_comparison.png', bbox_inches='tight', dpi=300)
plt.show()

# Save individual metric comparison plots
for metric in metrics:
    plt.figure(figsize=(6, 4))
    plt.bar(['ResNet18', 'PPNET'], [resnet18_metrics[metric], ppnet_metrics[metric]], 
            color=['steelblue', 'darkorange'], width=bar_width)
    plt.xlabel('Model', fontsize=12)
    plt.ylabel(metric.capitalize(), fontsize=12)
    plt.title(f'{metric.capitalize()} Comparison', fontsize=14)
    sns.despine()

    # Save each individual metric comparison plot
    plt.savefig(f'{metric}_comparison.png', bbox_inches='tight', dpi=300)
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set a consistent style for all plots
sns.set(style="whitegrid")

# Load the CSV files
resnet18_df = pd.read_csv('resnet18.csv')
ppnet_df = pd.read_csv('ppnet.csv')

# Extract the heart conditions (assuming the first column corresponds to heart conditions)
heart_conditions = resnet18_df.iloc[:, 0]

# Metrics to be compared
metrics = ['accuracy', 'auroc', 'f1_score']

# Plotting comparisons for each metric across all heart conditions
for metric in metrics:
    plt.figure(figsize=(12, 8))
    bar_width = 0.35
    x = range(len(heart_conditions))
    
    # Plot ResNet18 and PPNET metrics
    plt.bar(x, resnet18_df[metric], width=bar_width, label='ResNet18', color='steelblue')
    plt.bar([p + bar_width for p in x], ppnet_df[metric], width=bar_width, label='PPNET', color='darkorange')
    
    # Set labels and title
    plt.xlabel('Heart Conditions', fontsize=14)
    plt.ylabel(metric.capitalize(), fontsize=14)
    plt.title(f'Comparison of {metric.capitalize()} between ResNet18 and PPNET', fontsize=16)
    plt.xticks([p + bar_width / 2 for p in x], heart_conditions, rotation=45, ha="right")
    plt.legend(fontsize=12)

    # Remove top and right spines for a cleaner look
    sns.despine()

    # Save the plot as an image file
    plt.savefig(f'{metric}_comparison_per_condition.png', bbox_inches='tight', dpi=300)
    plt.show()
```
